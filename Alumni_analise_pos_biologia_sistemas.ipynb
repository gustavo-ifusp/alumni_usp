#**Q1. Sobre sua principal atuação atual, você se classifica como?**

1	 Sou Empregado em Instituição Privada<br>
2	 Sou Empregado em Instituição Pública ou Privada<br>
3	 Trabalho na minha área de formação<br>
4	 Trabalho Fora da minha área de formação<br>
5	 Sou Empreendedor<br>
6	 Sou Profissional Liberal<br>
7	 Sou Servidor Público Municipal<br>
8	 Sou Servidor Público Estadual<br>
9	 Sou Servidor Público Federal<br>
10	 Sou Pós-doutorando<br>
11	 Sou Estudante com Bolsa<br>
12	 Estou Procurando Emprego<br>
13	 Procuro meu Primeiro Emprego<br>
14	 Sou Estudante<br>
15	 Estou Aposentado<br>
15	 Estou Aposentado<br>
16	 Nenhuma das anteriores

```python
def ConverteNumIDresposta01Atua(id):
    dicionaio_resposta_01 = {
        1 : "Sou Empregado em Instituição Privada",
        2 : "Sou Empregado em Instituição Pública ou Privada",
        3 : "Trabalho na minha área de formação",
        4 : "Trabalho Fora da minha área de formação",
        5 : "Sou Empreendedor",
        6 : "Sou Profissional Liberal",
        7 : "Sou Servidor Público Municipal",
        8 : "Sou Servidor Público Estadual",
        9 : "Sou Servidor Público Federal",
        10 : "Sou Pós-doutorando",
        11 : "Sou Estudante com Bolsa",
        12 : "Estou Procurando Emprego",
        13 : "Procuro meu Primeiro Emprego",
        14 : "Sou Estudante",
        15 : "Estou Aposentado",
        15 : "Estou Aposentado",
        16 : "Nenhuma das anteriores"
                            }
    return dicionaio_resposta_01[id]
```

```python
resposta1 = dados["rpaqst1"]
```

```python
resposta1
```

```python
n = 0
pergunt1 = []
while n<23: #Quantidade de alunos que participaram da pesquisa 
  id = dados["rpaqst1"][n]
  perg1 = ConverteNumIDresposta01Atua(id)
  pergunt1.insert(n,perg1)
  n +=  1
print(pergunt1)
pergunta1 = pd.DataFrame()
pergunta1['Pergunta1'] = pergunt1
pergunta1
```

```python
sns.set(font_scale = 1.3)
g = sns.catplot(y='Pergunta1', data=pergunta1, kind="count", height=14,palette="husl", saturation=.5, order=pergunta1.Pergunta1.value_counts().index)
g.set(title = "Questionário Alumni - Pós em Biologia de Sistemas: Sobre sua principal atuação atual, você se classifica como?")
g.savefig("graf1_alumini_pos_bio_sistemas.png")
```

```python
fig = plt.figure(figsize=(12, 8))
colors = ['yellowgreen', 'red', 'gold', 'lightskyblue', 'white']
plt.pie(resposta1, labels=resposta1, startangle=90, autopct='%1.1f%%',pctdistance=0.8, textprops={'fontsize': 16})
plt.title('',  fontsize=28)
plt.axis('equal')
plt.show()
```
