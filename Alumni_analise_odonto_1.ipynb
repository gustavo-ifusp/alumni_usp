```python
df1.info()
```

```python
explode = (0.035, 0.035, 0.035)
fig = plt.figure(figsize=(12, 8))
plt.pie(df1['total'], labels=df1['Instituição'], explode=explode, shadow = True, startangle=90, autopct='%1.1f%%',pctdistance=0.5, textprops={'fontsize': 16})
plt.title('Total de inscritos das Faculdade de Odontologia no Alumni USP',  fontsize=28)
plt.axis('equal')
plt.show()
```

#**3. Status profissional dos inscritos das faculdades de  Odontologia na plataforma Alumni no período: 01/01/2016 até 04/10/2021**

image.png

```python
file = "status_profissional_odontologias.pdf"
df2 = pd.DataFrame()
df = tabula.read_pdf(file, pages = '1', multiple_tables = True)
df2 = df2.append(df, ignore_index=True)
```

```python
df2.columns
```

```python
df2 = df2.rename(columns={'Unnamed: 0': 'prof_status', 'Unnamed: 1':'2021','Unnamed: 2':'2020','Unnamed: 3':'2019','Unnamed: 4':'2018','Unnamed: 5':'2017', 'Unnamed: 6':'2016','Y / Record Count':'total_geral'})
df2 = df2[1:15]
```

```python
df2 = pd.DataFrame(df2)
df2.loc[df2['2021'] == '-', '2021'] = '0'
df2.loc[df2['2019'] == '-', '2019'] = '0'
df2.loc[df2['2017'] == '-', '2017'] = '0'
df2.loc[df2['2016'] == '-', '2016'] = '0'
df2.loc[df2['prof_status'] == 'Sou Pro ssional Liberal', 'prof_status'] = 'Sou Profissional Liberal'
df2['total_geral'] = df2['total_geral'].replace("1.433","1433")
```

```python
df2['2016'] = df2['2016'].astype('int')
df2['2021'] = df2['2021'].astype('int')
df2['2019'] = df2['2019'].astype('int')
df2['2017'] = df2['2017'].astype('int')
df2['2016'] = df2['2016'].astype('int')
df2['total_geral'] = df2['total_geral'].astype('int')
```

```python
df2 = df2.sort_values(by=['total_geral'], ascending=False)
df2
```

```python
sns.set(font_scale = 1.24)
g = sns.catplot(x="total_geral", y='prof_status', data=df2, kind="bar", height=14,saturation=.5)
g.set(title = "Status profissional Odontologias - 2016/2021")
g.savefig("prof_status_odontologias_multiplot1.png")
```

```python
df3 = df2.drop(['total_geral'], axis=1)
df4 = df3.melt('prof_status', var_name='cols',  value_name='vals')
g = sns.catplot(y='prof_status', x='vals', hue='cols', kind="bar", height=12,saturation=.5, data=df4)
g.set(title = "Status profissional Odontologias - 2016/2021")
sns.set(font_scale = 1.24)
g.savefig("prof_status_odontologias_multiplot2.png")
```
